```python
import pandas as pd
#Read Data
boston_df = pd.read_csv("https://raw.githubusercontent.com/vinay2k2/HealthcareBasicsAndImageProcessing/main/boston_train.csv",sep = ",")
boston_df.head()
```

```python
#Check if any numm value
boston_df.isnull()
```

```python
#Check if any numm value
boston_df.isnull().sum()
```

# Simple Plot
* Pass the column name to plt


```python
import matplotlib.pyplot as plt
plt.plot(boston_df['AGE'])
```

```python
boston_df['AGE'].describe()
```

# How to customize the figure size?


```python
from IPython.core.pylabtools import figsize
plt.figure(figsize=(20,8))
plt.plot(boston_df['AGE'])
```

# Can we plot everything together? Yes

```python
from IPython.core.pylabtools import figsize
plt.figure(figsize=(20,8))
plt.plot(boston_df)
```

# Now we don't know what is what? We can use Labels

```python
from IPython.core.pylabtools import figsize
plt.figure(figsize=(20,8))
plt.plot(boston_df['AGE'])
plt.xlabel("Row Index")
plt.ylabel("House Age")
plt.title("House Age Graph")
plt.show() # in jupyter notebook this is not important however for other editors we need to add this. 
#Also when you are trying to merge two different plots we need to use .show()
```

```python
# Lets plot Age and one more CRIM let say
from IPython.core.pylabtools import figsize
plt.figure(figsize=(20,8))
plt.plot(boston_df['AGE'])
plt.plot(boston_df['CRIM'])
plt.title("Age and CRIM")
plt.show() # in jupyter notebook this is not important however for other editors we need to add this. 
#Also when you are trying to merge two different plots we need to use .show()
```

# Sorting Values and Line Plot


```python
boston_df['AGE'].sort_values()
```

```python
from IPython.core.pylabtools import figsize
plt.figure(figsize=(20,8))
plt.plot(boston_df['AGE'].sort_values())
plt.xlabel("Row Index")
plt.ylabel("House Age")
plt.title("House Age Graph")
plt.show() 
# This looks bad so we can make the values unique to obtain a line chart
```

```python
from IPython.core.pylabtools import figsize
plt.figure(figsize=(20,8))
plt.plot(boston_df['AGE'].sort_values().unique())
plt.xlabel("Row Index")
plt.ylabel("House Age")
plt.title("House Age Graph")
plt.show() 
```

# For two Variables: Bar Plot

```python
plt.bar(x=boston_df['ZN'],height=boston_df['AGE'])
```

```python
#We can also specify color and width
plt.bar(x=boston_df['ZN'],height=boston_df['AGE'], width=1.6, color='blue')
```

```python
#Lets add more details
from IPython.core.pylabtools import figsize
plt.figure(figsize=(20,8))
plt.bar(x=boston_df['ZN'],height=boston_df['AGE'], width=1.6, color='blue')
plt.xlabel("ZN")
plt.ylabel("Age")
plt.title("Age vs ZN")
plt.show() 
```
